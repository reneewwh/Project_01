```python
# Initial imports
import pandas as pd
import numpy as np
import datetime as dt
from pathlib import Path

%matplotlib inline
```

```python
dt_p = Path("./nsw_property_data.csv")
```

```python
# Reading whale returns
pty_df = pd.read_csv(dt_p, parse_dates=True, infer_datetime_format=True)
pty_df.head()
```

```python
# check whether property id is unique values
display(pty_df['property_id'].duplicated().sum())
#check how many roles exist in data
display(pty_df.count())
```

```python
#remove property_name and strata_lot_number as they holds the majority of nulls and are irrelevant for scope of analysis
pty_df.drop(columns=['strata_lot_number', 'property_name','legal_description'],inplace=True)
pty_df
```

```python
pty_df['suburb']=pty_df['address'].str.rsplit(',').str[-1]
pty_df
```

```python
pty_df.isnull().sum()
```

```python
(pty_df['area'].isnull().sum()/pty_df['area'].isnull().count())*100
```

```python
#count number of records/ data size for analysis
pty_df.dropna().count()
```

```python
# To confirm sample size and commit
pty_df=pty_df.dropna().copy()
pty_df
```

```python
#set the right date format
pty_df['settlement_date'] = pd.to_datetime(pty_df['settlement_date'], format = '%Y-%m-%d')
#pty_df['contract_date'] = pd.to_datetime(pty_df['contract_date'], format = '%Y-%m-%d')
pty_df['download_date'] = pd.to_datetime(pty_df['download_date'], format = '%Y-%m-%d')
```

```python
pty_df['Settlement Month'] = pty_df['settlement_date'].dt.month
pty_df['Settlement Year'] = pty_df['settlement_date'].dt.year
```
